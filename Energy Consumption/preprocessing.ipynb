```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from tqdm import tqdm
```

```python
df = pd.read_csv('../data/BANES_cleaned.csv', parse_dates=['date'])
df.info()
```

```python
df = df.groupby(['date', 'location']).sum().reset_index()
df.info()
```

```python
df.drop(['totalunits'], axis=1, inplace=True)
df.describe()
```

```python
df['date'].min(), df['date'].max()
```

```python
days_range = pd.date_range(df['date'].min(), df['date'].max(), freq='D')
len(days_range)
```

```python
# locations where date is not in days_range
for local in df['location'].unique():
    if len(df[df['location'] == local]) == len(days_range):
        print(local, len(df[df['location'] == local]), len(days_range))
```
