```python
# group by day of the week
df_week = df.copy()
df_week.loc[:, 'day'] = pd.to_datetime(df_week['date']).dt.dayofweek

df_week.head()
```

```python
df_week.loc[:, 'is_weekend'] = df_week['day'].apply(lambda x: 1 if x in [5, 6] else 0)
df_week.head()
```

```python
# group by location and is_weekend
grouped_df_week = df_week.groupby(['location', 'is_weekend']).mean().iloc[:, 1:-2]
grouped_df_week.head()
```

```python
grouped_df_week.loc[:, 'mean'] = grouped_df_week.iloc[:, 1:].mean(axis=1)
grouped_df_week.head()
```

```python
grouped_df_week.reset_index(inplace=True)

weekend_df = grouped_df_week.loc[grouped_df_week['is_weekend'] == 1].loc[:, ['location', 'mean']]
weekday_df = grouped_df_week.loc[grouped_df_week['is_weekend'] == 0].loc[:, ['location', 'mean']]

weekend_df.head()
```

```python
# merge weekend and weekday dataframes
merged_df = pd.merge(weekend_df, weekday_df, on='location')
merged_df.head()
```

```python
# remove rows with mean_x or mean_y under 0
merged_df = merged_df.loc[merged_df['mean_x'] > 0]
merged_df = merged_df.loc[merged_df['mean_y'] > 0]
```

```python
# plot merged_df
plt.scatter(merged_df['mean_x'], merged_df['mean_y'], c="b", label="mean")
plt.plot([0, merged_df['mean_y'].max()], [0, merged_df['mean_y'].max()], 'r--', label="max")
plt.title('Mean kWh per location')
plt.xlabel('Mean kWh on weekends')
plt.ylabel('Mean kWh on weekdays')
plt.legend()
# make x axis and y axis the same limit
max_mean = max(merged_df['mean_x'].max(), merged_df['mean_y'].max())
plt.xlim(0, max_mean + 1)
plt.ylim(0, max_mean + 1)
plt.show()
```
