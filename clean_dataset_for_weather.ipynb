```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from tqdm import tqdm
```

```python
df_banes = pd.read_csv("BANES_cleaned.csv", parse_dates=["date"])
df_banes.head()
```

```python
df_banes.describe()
```

```python
ls = [df_banes] * 48
```

```python
# join ls into a single dataframe
df_banes_copy = pd.concat(ls, ignore_index=True)
```

```python
df_banes_copy.sort_values(by=["date"], inplace=True)
df_banes_copy.reset_index(drop=True, inplace=True)
df_banes_copy.head()
```

```python
df_banes_copy.shape, df_banes.shape
```

```python
columns = df_banes.columns[4:-2]
days_times = list(map(lambda x: pd.to_timedelta(x[1:].replace("_", ":") + ":00"), columns))
days_times *= df_banes.shape[0]
len(days_times), days_times[-5:]
```

```python
df_banes_copy.loc[:, 'day_time'] = days_times
df_banes_copy[['date', 'day_time']].tail()
```

```python
energy_cols = list(range(48)) * df_banes.shape[0]
len(energy_cols), energy_cols[-5:]
```

```python
# add energy_cols to df_banes_copy
df_banes_copy.loc[:, 'energy_col'] = energy_cols
df_banes_copy[['date', 'day_time', 'energy_col']].tail()
```

```python
df_banes_copy.loc[:, 'time'] = df_banes_copy['date'] + df_banes_copy['day_time']
df_banes_copy[['date', 'day_time', 'time', 'energy_col']].tail() 
```

```python
cols = [df_banes_copy.iloc[i, 4 + (i % 48)] for i in tqdm(range(df_banes_copy.shape[0]))]

# for i in tqdm(range(df_banes_copy.shape[0])):
#     val = df_banes_copy.iloc[i, 4 + (i % 48)]
#     cols.append(val)

df_banes_copy.loc[:, 'energy'] = np.array(cols)
df_banes_copy[['time', 'energy_col', 'energy']].head()
```

```python
df_banes_final = df_banes_copy[['time', 'location', 'postcode', 'energy']]
df_banes_final.head()
```

```python
df_banes_final.to_csv("BANES_cleaned_final.csv", index=False)
```
